## Task 2.2 
Assuming that the data collector makes an entry error when collecting data, it can be ensured that the error occurred in the 3P, 3PA and 3P% columns, but it is not sure which player's information the error lies on. Please try to explore the error by visualization to identify how many errors there are and try to fix it.


```python
# Code goes after this line by adding cells
```

```python
NBA_subset1 = NBA[['Player','3P', '3PA', '3P%']]
```

```python

NBA_subset1['correct_3P%'] = (NBA_subset1['3P'] /NBA_subset1['3PA'] )
```

```python
NBA_subset1['correct_3P%'] = NBA_subset1['correct_3P%'].round(3)
```

```python
NBA_subset2 = NBA_subset1.fillna(0) 
```

```python
NBA_subset2['status'] = NBA_subset2['3P%'] == NBA_subset2['correct_3P%']
```

```python
NBA_subset2
```

```python
NBA_subset2['status'].value_counts()
```

```python
sns.set_theme(style="darkgrid")
```

```python
freq_plot = sns.countplot(x='status', data=NBA_subset2)
plt.title('3P% Formula Match')
```

```python
NBA.dropna().boxplot(column='3P%')
```

```python
possible_errors = NBA.loc[(NBA['3P%'] < 0.1) | (NBA['3P%'] > 0.6) ]
```

```python
possible_errors['3P%']
```

```python
sns.histplot(data=possible_errors, x='3P%', bins=50)
plt.title('possible errors that are not zero')
```
